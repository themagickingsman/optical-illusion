# 🏛️ ARN Meroitic Data Ingestion Protocol

Use this notebook to download the critical Meroitic datasets required to expand the `Organic Codex` library.

### Sources:
1. **Meroitic Corpus (GitHub)**: Structured text and analysis data.
2. **Karanòg Inscriptions (Archive.org)**: 30+ Plates of high-resolution Meroitic Glyphs.


```python
# @title Step 1: Clone Meroitic Corpus (GitHub)
!git clone https://github.com/Joshua-Otten/Meroitic-Corpus.git
print("✅ Meroitic Corpus Cloned Successfully.")
```

```python
# @title Step 2: Download Karanog Inscriptions (Archive.org)
import os
import requests

# Create directory
os.makedirs('Karanog_Inscriptions', exist_ok=True)

# Direct PDF Link from Archive.org
pdf_url = "https://archive.org/download/karanmeroitici00grifuoft/karanmeroitici00grifuoft.pdf"
output_path = "Karanog_Inscriptions/Karanog_Meroitic_Inscriptions.pdf"

print("⬇️ Downloading Karanog PDF (This may take a moment).... ")
response = requests.get(pdf_url, stream=True)
if response.status_code == 200:
    with open(output_path, 'wb') as f:
        f.write(response.content)
    print(f"✅ Download Complete: {output_path}")
else:
    print("❌ Download Failed.")
```

```python
# @title Step 3: Bundle & Download
import shutil
from google.colab import files

# Zip the datasets
shutil.make_archive('meroitic_data_bundle', 'zip', '.')

print("📦 Bundle Created. Downloading to your machine...")
files.download('meroitic_data_bundle.zip')
```
